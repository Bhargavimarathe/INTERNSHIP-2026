```python
class Person:
    insti = "MITU"

    def __init__ (self, age, ht):
        self.age = age
        self.ht = ht

    def show(self):
        print(f"age -> {self.age}, height -> {self.ht}, institute -> {self.insti}")
```

```python
class Ravi(Person):

    def __init__(self, age, ht, sal):
        super().__init__(age, ht)
        self.sal = sal

    def show_sal(self):
        print(f"salary is {self.sal}")
```

```python
a = Ravi(25, 5.9, 57000)
```

```python
a.show_sal()
```

```python
a.show()
```

```python
# add items,view,exit(grocery list)
```

```python
class Menuapp:

    def __init__ (self):
        self.itemlist = []

    def show_menu(self):

        print("Menu\n1.add items\n2. view items\n3. exit")

    def add_items(self):
        item = input("enter the item name:")
        self.itemlist.append(item)

    def view_items(self):
        return self.itemlist

    def run(self):

        while True:

            self.show_menu()

            ch = int(input("enter the choice"))

            if ch == 1:
                self.add_items()
            elif ch == 2:
                self.view_items()
            elif ch == 3:
                break
            else:
                return 'invalid choice'
```

```python
atharv = Menuapp()
```

```python

atharv.run()
```

```python

atharv.itemlist
```

```python
# create a menu driven app for banking simulation
# check balance, deposit money, withdraw money (u can add more things as u see fit to make the app better)
```

```python
balance = 10000
transactions = []

while True:
    print("\n===== BANKING SYSTEM =====")
    print("1. Check Balance")
    print("2. Deposit Money")
    print("3. Withdraw Money")
    print("4. Transaction History")
    print("5. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        print(f"Current Balance: ₹{balance}")

    elif choice == 2:
        amount = float(input("Enter amount to deposit: ₹"))
        if amount > 0:
            balance += amount
            transactions.append(f"Deposited ₹{amount}")
            print("Money deposited successfully!")
        else:
            print("Invalid amount!")

    elif choice == 3:
        amount = float(input("Enter amount to withdraw: ₹"))
        if amount <= balance:
            balance -= amount
            transactions.append(f"Withdrawn ₹{amount}")
            print("Money withdrawn successfully!")
        else:
            print("Insufficient balance!")

    elif choice == 4:
        print("\n--- Transaction History ---")
        if transactions:
            for t in transactions:
                print(t)
        else:
            print("No transactions yet.")

    elif choice == 5:
        print("Thank you for using the Banking System!")
        break

    else:
        print("Invalid choice! Please try again.")
```
