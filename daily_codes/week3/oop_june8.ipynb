```python
"""
create a class called employee=>
attributes -> company,name,salary,job_designation
here all the objects of this company will have the same common company
make salary as private variable

create 3 objects-> e1,e2,e3, with given attributes
method-> 1, method to calculate the annual salary
         2. method to show entire data

""";
```

```python
class Employee:
    company = "TCS"

    def __init__(self, name, salary, job_designation):
        self.name = name
        self.__salary = salary
        self.job_designation = job_designation

    def annual_salary(self):
        return self.__salary * 12

    def show_data(self):
        print("Company:", Employee.company)
        print("Name:", self.name)
        print("Salary:", self.__salary)
        print("Job Designation:", self.job_designation)
        print("Annual Salary:", self.annual_salary())


e1 = Employee("Riya", 50000, "Data Analyst")
e2 = Employee("Amit", 60000, "Python Developer")
e3 = Employee("Sneha", 70000, "Project Manager")

e1.show_data()
e2.show_data()
e3.show_data()
```

```python
 #2nd method
 class Employee:
    company = "TCS"

    def __init__(self, name, salary, job_designation):
        self.name = name
        self.__salary = salary
        self.job_designation = job_designation

    def annual_salary(self):
        return self.__salary * 12

    def info(self):
        return f"Name={self.name}, Company={Employee.company}, Salary={self.__salary}, Job_Designation={self.job_designation}, Annual_Salary={self.annual_salary()}"

```
