```python
# creatr a function to find the factoria of a number
```

```python
def fact(n):
  fact= 1
  for i in range(1,n+1):
    fact= fact*i
  print(fact)
```

```python
fact(5)
```

```python
fact(10)
```

```python
# type hinting => tell the type of the parameter and return  value, and docstring
```

```python
def sq(x: int)-> int:

  """
  Calculate the square of the number

  Args:
    x(int): the number to square

  Returns:
     int: the square of the number
  """
  res=x*x
  return res
```

```python
help(sq)
```

```python
sq(5)
```
