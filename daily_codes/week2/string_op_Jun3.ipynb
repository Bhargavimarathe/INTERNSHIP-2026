```python
"""
ytake a string sas user input
use a loop and if-else to check if the specific character is present in the striong or not
get output for both if the character id present and if is ot present
take 2 characters to check in the string

user unpit -> string, character1, character2
""";
```

```python
str = input("enter the string ")
char1 = input("enter the char 1 to find ")
char2 = input("enter the char 2 to find ")

found_char1 = False
found_char2 = False

for char in str:
    if char == char1:
        found_char1 = True
    if char == char2:
        found_char2 = True

if found_char1:
    print(f"the character '{char1}' is present in the string")
else:
    print(f"the character '{char1}' is not present in the string")

if found_char2:
    print(f"the character '{char2}' is present in the string")
else:
    print(f"the character '{char2}' is not present in the string")
```
