```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'roll': [1, 2, 3, 4, 5, 6, 7, 8, 9, 9, 10, 11, 12, 13, 14, 15],
    'name': ['anil', 'amit', 'aniket', 'ajinkya', 'asha', 'ayesha',
             'amar', 'Amita', 'amol', 'amol', 'anmol', 'surbhi',
             'Sujay', 'subhash', 'sachin', 'shalini'],
    'class': ['TE', 'TE', 'BE', 'TE', 'TE', 'BE',
              'BE', 'be', 'TE', 'TE', 'BE', 'TE',
              'TE', 'be', np.nan, 'TE'],
    'marks': [56.77, 59.77, 76.88, 69.66, 63.28, np.nan,
              65.34, 68.33, 56.75, 56.75, 78.66, np.nan,
              74.12, 88.45, 59.33, 63.45],
    'age': [22.0, 21.0, 19.0, 20.0, np.nan, 20.0,
            19.0, 23.0, 20.0, 20.0, np.nan, 23.0,
            22.0, 21.0, 21.0, 20.0]
})

```

```python
df['marks'].fillna(45)
```

```python
# fill with avg marks
```

```python
df['marks'].fillna(df['marks'].mean())
```

```python
# fill age with median
```

```python
df['age'].fillna(df['age'].median())
```

```python
df['class'].str.upper()
```

```python
df['class'].fillna(df['class'].mode()[0])
```
