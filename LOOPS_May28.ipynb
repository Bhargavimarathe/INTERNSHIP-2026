```python
# find the addition of odd numbers between 1 to 20
```

```python
sum = 0
i = 1

while i <= 20:
    if i % 2 != 0:
        sum = sum + i
    i = i + 1

print("Addition =", sum)
```

```python
# write a program to find factorial of a number  using while loop
```

```python
# using increasing
num = int(input("Enter a number: "))

fact = 1
i = 1

while i <= num:
    fact = fact * i
    i = i + 1

print("Factorial =", fact)
```

```python
# using decresing
num = int(input("Enter a number: "))

fact = 1

while num > 0:
    fact = fact * num
    num -= 1

print("Factorial =", fact)
```

```python
# find the addition of digits
# ed-> 12345=> 1+2+3+4+5=15
# make the use of % and //
```

```python
num = int(input("Enter a number: "))
sum = 0

while num > 0:
    digit = num % 10
    sum = sum + digit
    num = num // 10

print("Addition =", sum)
```

```python
# factorial using for loop
```

```python
num = int(input("Enter a number: "))
fact = 1

for i in range(1, num + 1):
    fact = fact * i

print("Factorial =", fact)
```
