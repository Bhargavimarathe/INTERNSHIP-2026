```python
"""
Question 1

Download the Seeds dataset from UCI Repository and perform the following operations:

Read the dataset and separate input and output columns.
Plot the graph of class balancing.
Perform train-test split with 80% and 20%.
Create Logistic Regression model.
Find Accuracy, Confusion Matrix and Classification Report.
Predict a single value.
""";
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
column_names = ['Area','Perimeter','Compactness','Kernel_Length',
                'Kernel_Width','Asymmetry_Coefficient',
                'Kernel_Groove_Length','Class']
```

```python
df = pd.read_csv("seeds_dataset.txt", sep=r"\s+", names=column_names)
```

```python
df.head()
```

```python
# Separate Input and Output
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
# Plot Class Balance
sns.countplot(x=y)
plt.title("Class Balance")
plt.show()
```

```python
# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.20,
    random_state=42,
    stratify=y
)
```

```python
# Logistic Regression Model
model = LogisticRegression(max_iter=500)
model.fit(X_train, y_train)
```

```python
# Prediction
y_pred = model.predict(X_test)
```

```python
# Accuracy
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
sns.heatmap(cm, annot=True, cmap="Blues", fmt="d")
plt.title("Confusion Matrix")
plt.show()
```

```python
# Classification Report
print(classification_report(y_test, y_pred))
```

```python
# Single Prediction
sample = [[15.26,14.84,0.8710,5.763,3.312,2.221,5.220]]

prediction = model.predict(sample)

print("Predicted Class:", prediction)
```

```python
"""
Problem Statement 2: Balance Scale Dataset (Decision Tree)
Question 2

Download the Balance Scale Dataset and perform:

Read dataset.
Separate input/output columns.
Plot class balancing graph.
Train-test split (75%-25%).
Create Decision Tree model.
Find Accuracy, Confusion Matrix and Classification Report.
Predict single value.
Plot and save the Decision Tree.
""";
```

```python

from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
# Load Dataset
column_names = ['Class','Left_Weight','Left_Distance',
                'Right_Weight','Right_Distance']
```

```python
df = pd.read_csv("balance-scale.data", names=column_names)
```

```python
df.head()
```

```python

# Separate Input and Output
X = df.iloc[:,1:]
y = df.iloc[:,0]
```

```python
# Class Balance
sns.countplot(x=y)
plt.title("Class Balance")
plt.show()
```

```python

# Train Test Split
X_train, X_test, y_train, y_test = train_test_split(
    X,y,
    test_size=0.25,
    random_state=42,
    stratify=y
)
```

```python

# Decision Tree Model
model = DecisionTreeClassifier(random_state=42)
```

```python
model.fit(X_train,y_train)
```

```python
# Prediction
y_pred = model.predict(X_test)
```

```python
# Accuracy
print("Accuracy:",accuracy_score(y_test,y_pred))
```

```python
# Confusion Matrix
cm = confusion_matrix(y_test,y_pred)

print(cm)

sns.heatmap(cm,annot=True,cmap="Blues",fmt="d")
plt.show()
```

```python
# Classification Report
print(classification_report(y_test,y_pred))
```

```python
# Single Prediction
sample=[[3,5,2,4]]

prediction=model.predict(sample)

print("Prediction:",prediction)
```

```python
# Plot Tree
plt.figure(figsize=(20,10))

plot_tree(model,
          feature_names=X.columns,
          class_names=model.classes_,
          filled=True)

plt.savefig("decision_tree_balance_scale.png")

plt.show()
```

```python
"""
Problem Statement 3: Computer Hardware Dataset (Decision Tree Regression)
Question 3

Download the Computer Hardware Dataset and perform:

Read dataset.
Separate input/output.
Train-test split (75%-25%).
Create Decision Tree model.
Find Mean Absolute Error and Accuracy.
Predict single value.
Plot and save tree.
""";
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor, plot_tree

from sklearn.metrics import mean_absolute_error, r2_score
```

```python

# Load Dataset
column_names=[
'Vendor','Model','MYCT','MMIN','MMAX','CACH',
'CHMIN','CHMAX','PRP','ERP'
]
```

```python

df=pd.read_csv("machine.data",names=column_names)
```

```python
df.head()
```

```python
# Convert Categorical Columns
df=pd.get_dummies(df,columns=['Vendor','Model'])
```

```python
# Input Output
X=df.drop(['PRP','ERP'],axis=1)
y=df['PRP']
```

```python
# Train Test Split
X_train,X_test,y_train,y_test=train_test_split(
    X,y,
    test_size=0.25,
    random_state=42
)
```

```python
# Decision Tree Regressor
model=DecisionTreeRegressor(random_state=42)

model.fit(X_train,y_train)
```

```python
# Prediction
y_pred=model.predict(X_test)
```

```python
# Mean Absolute Error
print("MAE:",mean_absolute_error(y_test,y_pred))
```

```python

# R2 Score
print("R2 Score:",r2_score(y_test,y_pred))
```

```python
# Single Prediction
sample=[X.iloc[0].tolist()]

prediction=model.predict(sample)

print("Prediction:",prediction)
```

```python
# Plot Tree
plt.figure(figsize=(25,10))

plot_tree(
    model,
    feature_names=X.columns,
    filled=True,
    max_depth=3
)
plt.savefig("computer_hardware_tree.png")

plt.show()
```
