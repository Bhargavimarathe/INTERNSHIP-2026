```python
#Write a function analyze_sentence(sentence) that takes a sentence and returns a dictionary with:
#word_count,char_count (excluding spaces),unique_words (case-insensitive),most_frequent_word
#Test it with: "The quick brown fox jumps over the lazy dog the fox"

def analyze_sentence(sentence):
    words = sentence.lower().split()

    result = {
        "word_count": len(words),
        "char_count": len(sentence.replace(" ", "")),
        "unique_words": len(set(words)),
        "most_frequent_word": max(set(words), key=words.count)
    }

    return result

text = "Python is easy and python is powerful and python is fun"

print(analyze_sentence(text))
```

```python
# Given the string:
#text = "A man a plan a canal Panama"
#Do the following and print each result:
#a) Reverse the string without using slicing.
#b) Check if it is a palindrome, ignoring case and spaces.
#c) Replace every vowel with *.
#d) Count how many times each vowel appears.

text = "A man a plan a canal Panama"

rev = ""
for ch in text:
    rev = ch + rev
print("Reversed String:", rev)

s = text.lower().replace(" ", "")

if s == "".join(reversed(s)):
    print("Palindrome")
else:
    print("Not Palindrome")

new_text = ""

for ch in text:
    if ch.lower() in "aeiou":
        new_text += "*"
    else:
        new_text += ch

print("After replacing vowels:", new_text)

vowels = {'a': 0, 'e': 0, 'i': 0, 'o': 0, 'u': 0}

for ch in text.lower():
    if ch in vowels:
        vowels[ch] += 1

print("Vowel Count:", vowels)
```

```python
#Write a function format_report(data) that takes a list of tuples (name, score, grade) and prints a formatted table where:
#name is left-aligned (width 15)
#score is right-aligned (width 6)
#grade is centered (width 8)

def format_report(data):
    print(f"{'Name':<15}{'Score':>6}{'Grade':^8}")

    for name, score, grade in data:
        print(f"{name:<15}{score:>6}{grade:^8}")

data = [
    ("Bhargavi", 92, "A"),
    ("Sakshi", 78, "B+"),
    ("Prachi", 55, "C"),
    ("Anushri", 100, "A+")
]

format_report(data)
```

```python
# Write a function list_ops(nums) that:
#a) Removes duplicates while preserving the original order.
#b) Returns a new list with only elements at even indices, squared.
#c) Flattens the nested list into a single list using list comprehension.
def list_ops(nums):
    unique = []

    for i in nums:
        if i not in unique:
            unique.append(i)

    even_square = [nums[i]**2 for i in range(0, len(nums), 2)]

    nested = [
        [1, 2, 3],
        [4, 5],
        [6, 7, 8, 9]
    ]

    flat = [j for i in nested for j in i]

    print("Without Duplicates:", unique)
    print("Even Index Squared:", even_square)
    print("Flattened List:", flat)

nums = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5]

list_ops(nums)
```

```python
#Write a Python function power_all(*args, exp=2) that accepts any number of arguments and returns a
#list containing each number raised to the power specified by exp. The default value of exp should be 2.
#b) Write a Python function profile(**kwargs) that accepts any number of keyword arguments and prints each key-value pair in the format:
#Key : Value
#c) Given a list of dictionaries containing student names and CGPA values, use a lambda function inside the sorted()
 #function to sort the students in descending order of CGPA and print the sorted list.

def power_all(*args, exp=2):
    result = []

    for i in args:
        result.append(i ** exp)

    return result


def profile(**kwargs):
    for key, value in kwargs.items():
        print(key, ":", value)


students = [
    {"name": "Arjun", "cgpa": 8.4},
    {"name": "Sanya", "cgpa": 9.1},
    {"name": "Dev", "cgpa": 7.8}
]

print("Power Function:")
print(power_all(2, 3, 4))

print("\nProfile Details:")
profile(name="Bhargavi", age=20, city="Pune")

print("\nStudents Sorted by CGPA:")

result = sorted(students, key=lambda x: x["cgpa"], reverse=True)

for i in result:
    print(i)
```

```python
#Create a module stats_utils.py with three functions:
#a) mean(data) — arithmetic mean
#b) variance(data) — population variance without using the statistics module
#c) normalize(data) — min-max normalization returning values between 0 and 1
#Use this dataset:
#dataset = [12, 45, 7, 23, 56, 89, 34, 67, 11, 90]
#In a separate main.py, import and use these functions. Also compute the standard deviation using the math module.

%%writefile stats_utils.py

def mean(data):
    return sum(data) / len(data)

def variance(data):
    m = mean(data)
    total = 0

    for i in data:
        total += (i - m) ** 2

    return total / len(data)

def normalize(data):
    mn = min(data)
    mx = max(data)

    result = []

    for i in data:
        result.append((i - mn) / (mx - mn))

    return result
```

```python
import math
from stats_utils import mean, variance, normalize

dataset = [12, 45, 7, 23, 56, 89, 34, 67, 11, 90]

print("Mean =", mean(dataset))
print("Variance =", variance(dataset))
print("Normalized Data =", normalize(dataset))

std_dev = math.sqrt(variance(dataset))
print("Standard Deviation =", std_dev)
```

```python
#Complete the following:
#a) Write a closure make_multiplier(n) that returns a function which multiplies its input by n.
 #Use this closure to create a triple function and a quintuple function, and demonstrate their usage.
#b) Write a decorator @timer that measures and prints the execution time of any function it wraps.
 #Apply this decorator to a function slow_sum(n) that computes the sum of squares from 1 to n.
#c) Write a decorator @validate_positive that checks whether all arguments passed to a function are positive numbers.
 #If any argument is zero or negative, it should raise a ValueError.
  #Apply this decorator to a function safe_sqrt(x) that returns the square root of a number.

import time
import math

def make_multiplier(n):
    def multiply(x):
        return x * n
    return multiply

triple = make_multiplier(3)
quintuple = make_multiplier(5)

print("Triple of 10:", triple(10))
print("Quintuple of 10:", quintuple(10))

def timer(func):
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        end = time.time()
        print("Execution Time:", end - start, "seconds")
        return result
    return wrapper

@timer
def slow_sum(n):
    total = 0
    for i in range(1, n + 1):
        total += i ** 2
    return total

print("Sum of Squares:", slow_sum(100000))

def validate_positive(func):
    def wrapper(*args, **kwargs):
        for arg in args:
            if arg <= 0:
                raise ValueError("All arguments must be positive numbers")
        return func(*args, **kwargs)
    return wrapper

@validate_positive
def safe_sqrt(x):
    return math.sqrt(x)

print("Square Root:", safe_sqrt(25))
```
