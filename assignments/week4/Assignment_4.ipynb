```python
"""
Write a function safe_divide(a, b) that divides two numbers and handles ZeroDivisionError
and TypeError separately, printing an appropriate message for each. Use a finally block to print 'Operation attempted.'
 regardless of the outcome. Test it with valid numbers, division by zero, and a string input.
""";
```

```python
def safe_divide(a, b):
    try:
        result = a / b
        print("Result:", result)

    except ZeroDivisionError:
        print("Error: Cannot divide by zero.")

    except TypeError:
        print("Error: Invalid data type. Please enter numbers only.")

    finally:
        print("Operation attempted.")


print("Test 1: Valid numbers")
safe_divide(10, 2)

print("\nTest 2: Division by zero")
safe_divide(10, 0)

print("\nTest 3: String input")
safe_divide(10, "5")
```

```python
"""
3

Write a program that repeatedly asks the user (use a loop with at least 3 simulated inputs in a list, e.g. user_inputs = ['10', 'abc', '0', '5']) to enter a number and divide 100 by it. Handle the following in a single try block using multiple except clauses:

a) ValueError if the input is not a number.

b) ZeroDivisionError if the input is zero.

c) A generic Exception as a catch-all for anything else.

Print the result if successful, otherwise print which error occurred.
""";
```

```python
user_inputs = ['10', 'abc', '0', '5']

for value in user_inputs:
    try:
        num = int(value)
        print("Result:", 100 / num)

    except ValueError:
        print("Error: Input is not a number.")

    except ZeroDivisionError:
        print("Error: Cannot divide by zero.")

    except Exception:
        print("Error: Something went wrong.")
```

```python
"""
Generate a NumPy array of 15 random integers between 1 and 100 using np.random.randint() (set a seed of 42 for reproducibility). Then:

a) Find the mean, median, standard deviation, min, and max.

b) Sort the array in descending order.

c) Find the indices of the 3 largest values using np.argsort().

d) Normalize the array so all values lie between 0 and 1.
""";
```

```python
import numpy as np

np.random.seed(42)

arr = np.random.randint(1, 101, 15)

print("Array:", arr)

print("Mean:", np.mean(arr))
print("Median:", np.median(arr))
print("Standard Deviation:", np.std(arr))
print("Minimum:", np.min(arr))
print("Maximum:", np.max(arr))

print("Descending Order:", np.sort(arr)[::-1])

print("Indices of 3 Largest Values:", np.argsort(arr)[-3:])

normalized = (arr - np.min(arr)) / (np.max(arr) - np.min(arr))
print("Normalized Array:", normalized)
```

```python
"""
7.

Create a Pandas DataFrame from the dictionary below and perform the following:

data = {
    "Name": ["Aarav", "Priya", "Karan", "Meera", "Sahil"],
    "Age": [23, 21, 25, 22, 24],
    "Department": ["AI", "Data Science", "AI", "Web Dev", "Data Science"],
    "Score": [88, 92, 79, 95, 67]
}

a) Display the first 3 rows and the data types of each column.

b) Add a new column Grade based on Score (>=90 'A', >=75 'B', else 'C').

c) Sort the DataFrame by Score in descending order.

d) Filter rows where Department is 'AI'.
""";
```

```python
import pandas as pd

data = {
    "Name": ["Aarav", "Priya", "Karan", "Meera", "Sahil"],
    "Age": [23, 21, 25, 22, 24],
    "Department": ["AI", "Data Science", "AI", "Web Dev", "Data Science"],
    "Score": [88, 92, 79, 95, 67]
}

df = pd.DataFrame(data)

print("First 3 Rows:")
print(df.head(3))

print("\nData Types:")
print(df.dtypes)

df["Grade"] = df["Score"].apply(
    lambda x: "A" if x >= 90 else "B" if x >= 75 else "C"
)

print("\nDataFrame with Grade:")
print(df)

print("\nSorted by Score (Descending):")
print(df.sort_values(by="Score", ascending=False))

print("\nStudents from AI Department:")
print(df[df["Department"] == "AI"])
```

```python
"""
8.

Using the same DataFrame from Q7:

a) Group by Department and find the average Score for each.

b) Find the row with the highest Score using idxmax().

c) Rename the column Score to Marks.

d) Drop the Age column and display the resulting DataFrame.
""";
```

```python
import pandas as pd

data = {
    "Name": ["Aarav", "Priya", "Karan", "Meera", "Sahil"],
    "Age": [23, 21, 25, 22, 24],
    "Department": ["AI", "Data Science", "AI", "Web Dev", "Data Science"],
    "Score": [88, 92, 79, 95, 67]
}

df = pd.DataFrame(data)

print("Average Score by Department:")
print(df.groupby("Department")["Score"].mean())

print("\nRow with Highest Score:")
print(df.loc[df["Score"].idxmax()])

df.rename(columns={"Score": "Marks"}, inplace=True)

print("\nDataFrame after Renaming Score to Marks:")
print(df)

df.drop("Age", axis=1, inplace=True)

print("\nDataFrame after Dropping Age Column:")
print(df)
```

```python
"""
9.

Create a DataFrame with some missing values as shown below and handle them:

import numpy as np

data = {
    "Product": ["Pen", "Notebook", "Eraser", "Sharpener", "Ruler"],
    "Price": [10, np.nan, 5, 8, np.nan],
    "Stock": [100, 50, np.nan, 30, 20]
}

a) Check the number of missing values in each column using isnull().sum().

b) Fill missing Price values with the column mean.

c) Drop rows where Stock is still missing.

d) Save the cleaned DataFrame to a CSV file named cleaned_products.csv without the index.
""";
```

```python
import pandas as pd
import numpy as np

data = {
    "Product": ["Pen", "Notebook", "Eraser", "Sharpener", "Ruler"],
    "Price": [10, np.nan, 5, 8, np.nan],
    "Stock": [100, 50, np.nan, 30, 20]
}

df = pd.DataFrame(data)

print("Missing Values in Each Column:")
print(df.isnull().sum())

df["Price"].fillna(df["Price"].mean(), inplace=True)

print("\nDataFrame after Filling Missing Price Values:")
print(df)

df = df.dropna(subset=["Stock"])

print("\nDataFrame after Dropping Rows with Missing Stock:")
print(df)

df.to_csv("cleaned_products.csv", index=False)

print("\nData saved to cleaned_products.csv")
```
